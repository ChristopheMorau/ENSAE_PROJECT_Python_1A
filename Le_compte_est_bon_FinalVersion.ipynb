```python
#FIRST VERSION: quicker but incomplete version. Does not do operations like a/(c+d) or (a+b)*(c+d) or (a*b)+(c*d) etc.

n=int(input('How many integers do you want to use? NB: more than 6 can take a long period of time')) 

import random
from random import randint

def random_draw() :
    set1 = [k for k in range(1,11)]
    set2 = [25*k for k in range(1,5)]
    data = set1 + set1 + set2 + set2
    random.shuffle(data)
    return data[0:n]

target_number = randint(100,999)
draw = random_draw()


print('We can use',draw)
print('We want :',target_number)
print(' ')

class Node :
    def __init__(self, value, available, children = [], ancester=[]) :
        self.value = value
        self.available = available
        self.children = children
        self.ancester = ancester

class Tree:
    def __init__(self,available=draw):
        build_children(Node(0,available))
        self.tree = Node(0,available)
      

    
def ancesters(node):
    genealogy = [node]
    if node.ancester is not None:
        daddy = node.ancester
    while daddy.value != 0:
        genealogy.append(daddy)
        daddy=daddy.ancester
    return genealogy[::-1]

def build_children(node) :
    if node.available == [0 for i in range(len(node.available))] :
        return 
    for k in range(len(node.available)) :
        if node.available[k] != 0 :
            copy = node.available.copy()
            copy[k] = 0
            node.children.append(Node(node.value + node.available[k], copy, [], node))    
            if node.value != 0:
                node.children.append(Node(node.value * node.available[k], copy, [], node))
            if node.value-node.available[k] > 0 :
                node.children.append(Node(node.value - node.available[k], copy, [], node))
            div = node.value / node.available[k]
            if div == int(div) and div >1:
                node.children.append(Node(node.value / node.available[k], copy, [], node))
    for children in node.children:
        build_children(children)

Q=Tree()


def find_target(node,target,lengths=[],paths=[], values_knowledge = []):
    if node is None:
        return
    if node.value == target:
        ancesters_values = [ancester.value for ancester in ancesters(node)[1:]]
        if set(ancesters_values) not in values_knowledge :
            values_knowledge.append(set(ancesters_values))
            paths.append(ancesters(node))
            lengths.append(len(ancesters(node)))
    for children in node.children :
        find_target(children,target, lengths, paths, values_knowledge)
    return lengths, paths

[lengths, paths] = find_target(Q.tree, target_number)


def shortest_paths(lengths, paths):
    positions=[]
    if len(lengths)>0:
        m=min(lengths)
        for k in range(len(lengths)):
            if lengths[k]==m:
                positions.append(k)
    return [paths[i] for i in positions]

shortest_paths = shortest_paths(lengths,paths)

def missing_coordinate(vector1,vector2):
    for k in range(len(vector1)):
        if vector1[k] != vector2[k]:
            return vector1[k]

def path_clarifier(path, target_number):
    n=len(path)
    for j in range(n-1):
        used_number = missing_coordinate(path[j].available, path[j+1].available)
        if path[j+1].value == path[j].value + used_number:
            print(path[j].value,'+',used_number,'=',path[j+1].value)
        if path[j+1].value == path[j].value - used_number:
            print(path[j].value,'-',used_number,'=',path[j+1].value)
        if path[j+1].value == path[j].value * used_number:
            print(path[j].value,'x',used_number,'=',path[j+1].value)
        if path[j+1].value == path[j].value / used_number:
            print(path[j].value,'/',used_number,'=',path[j+1].value)




if paths == []:
    print('THERE ARE NO SOLUTION')
else:
    print('THE OPTIMAL SOLUTIONS ARE :')
    for path in shortest_paths:
        path_clarifier(path,target_number)
        print('----')
    
    print('THE SOLUTIONS ARE:')
    for path in paths:
        path_clarifier(path,target_number)
        print('----')
```

```python
#Second version: Longer one, to be used in case the first version finds no solution 
#or if you want to have every single solution (cf cell below) using operations like a+(b+c)/(d+e) or a*b+c*d etc

#Part 1: to check if we really have no solution:

class Node :
    def __init__(self, memo, children = [], track = []) :
        self.memo = memo
        self.children = children
        self.track = track

class Tree :
    def __init__(self, memo = draw, target_number = target_number) :
        build_children(Node(memo), target_number)
        self.tree = Node(memo)

def build_children(node, target_number) :
    if len(node.memo) == 1 :
            return
    if target_number in node.memo :
        return
    the_Knowledge = []
    length = len(node.memo)
    for k in range(length) :
        if node.memo[k] != 0 :
            for j in range(length) :
                if j != k :
                    if node.memo[j] != 0 :
                        [[copy1, track1], [copy2, track2], [copy3, track3], [copy4, track4]] = [[node.memo.copy(), node.track.copy()] for k in range(4)]
                        Sum = node.memo[k] + node.memo[j]
                        Diff = node.memo[k] - node.memo[j]
                        Prod = node.memo[k] * node.memo[j]
                        Div = node.memo[k] / node.memo[j]
                        if j != 0 :
                            if set([node.memo[k], '+', node.memo[j]]) not in the_Knowledge :    
                                copy1[k] = copy1[0]
                                copy1[j] = Sum
                                track1.append([node.memo[k], '+', node.memo[j], '=', Sum])
                                node.children.append(Node(copy1[1:], [], track1))
                                the_Knowledge += [set([node.memo[k], '+', node.memo[j]])]
                            if Diff > 0 :
                                copy2[k] = copy2[0]
                                copy2[j] = Diff
                                track2.append([node.memo[k], '-', node.memo[j], '=', Diff])
                                node.children.append(Node(copy2[1:], [], track2))
                            if set([node.memo[k], 'x', node.memo[j]]) not in the_Knowledge :    
                                copy3[k] = copy3[0]
                                copy3[j] = Prod
                                track3.append([node.memo[k], 'x', node.memo[j], '=', Prod])
                                node.children.append(Node(copy3[1:], [], track3))
                                the_Knowledge += [set([node.memo[k], 'x', node.memo[j]])]

                            if Div == int(Div) :
                                copy4[k] = copy4[0]
                                copy4[j] = Div
                                track4.append([node.memo[k], '/', node.memo[j], '=', Div])
                                node.children.append(Node(copy4[1:], [], track4))

                        if j == 0 :
                            if set([node.memo[k], '+', node.memo[j]]) not in the_Knowledge :    
                                copy1[k] = Sum
                                track1.append([node.memo[k], '+', node.memo[j], '=', Sum])
                                node.children.append(Node(copy1[1:], [], track1))
                                the_Knowledge += [set([node.memo[k], '+', node.memo[j]])]

                            if Diff > 0 :
                                copy2[k] = Diff
                                track2.append([node.memo[k], '-', node.memo[j], '=', Diff])
                                node.children.append(Node(copy2[1:], [], track2))

                            if set([node.memo[k], 'x', node.memo[j]]) not in the_Knowledge:
                                copy3[k] = Prod
                                track3.append([node.memo[k], 'x', node.memo[j], '=', Prod])
                                node.children.append(Node(copy3[1:], [], track3))
                                the_Knowledge += [set([node.memo[k], 'x', node.memo[j]])]

                            if Div == int(Div) :
                                copy4[k] = Div
                                track4.append([node.memo[k], '/', node.memo[j], '=', Div])
                                node.children.append(Node(copy4[1:], [], track4))
                                the_Knowledge += [set([node.memo[k], '/', node.memo[j]])]
                            


    for children in node.children :
        build_children(children, target_number)

T = Tree()

#OPTIMAL SEARCH

results = []
lengths = []

def stack_builder(target, node, results, lengths) :
    if target in node.memo :
        results.append(node.track)
        lengths.append(len(node.track))
    for children in node.children :
        stack_builder(target_number, children, results, lengths)

stack_builder(target_number, T.tree, results, lengths)


print('We can use:',draw)
print('We want :',target_number)
print(' ')

if lengths == []:
    print('THERE ARE NO SOLUTION')

optimal_paths = []
if len(lengths) !=0:
    m = min(lengths)
    for path in results :
        if len(path) == m:
            optimal_paths.append(path)
                
    print('THE OPTIMAL SOLUTIONS ARE:')
    for path in optimal_paths:
        for k in range(len(path)):
            print(path[k][0],path[k][1],path[k][2],path[k][3],path[k][4])
        print('----')
```
